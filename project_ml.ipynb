### Creating Regression Tree
we can create regression tree with the help of classes there are two nodes in the class 


1.   Internal node : contains information on which parameter we need to split and at what point
2.   Leaf Node: contain avg value



```python
class TreeNode():
    def __init__(self, feature_index=None, threshold=None, left=None, right=None, varince_reduction=None, value=None):#setting default argument None
        
        # for decision node
        self.feature_index = feature_index # for storing feature on which we need to split
        self.threshold = threshold # storing threshold value on which we need to deside
        self.left = left # storing left children < threshold
        self.right = right # storing right children > threshold
        self.varince_reduction = varince_reduction
        
        # for leaf node
        self.value = value
    

```

```python
class Decision_Tree_regression():
    def __init__(self, min_split_sample=2, max_depth=2):
        # condition for where to stop
        self.max_depth = max_depth
        self.min_split_sample = min_split_sample
        self.root = None # initialize the root of the tree
  
    def fit(self, X, Y):
        #function to train the tree
        
        dset = np.concatenate((X, Y), axis=1)
        self.root = self.make_tree(dset)
        
    def make_tree(self, dset, curr_depth=0):
        #used recursive function for building tree
        
        X, Y = dset[:,:-1], dset[:,-1]
        num_samples, num_features = np.shape(X)
        best_split = {}
        # split until stopping conditions are met
        if num_samples>=self.min_split_sample and curr_depth<=self.max_depth:
            # find the best split
            best_split = self.get_best_split(dset, num_samples, num_features)
            # check if information gain is positive
            if best_split["varince_reduction"]>0:
                # recur left
                left_subtree = self.make_tree(best_split["dataset_left"], curr_depth+1)
                # recur right
                right_subtree = self.make_tree(best_split["dataset_right"], curr_depth+1)
                # return decision node
                return TreeNode(best_split["feature_index"], best_split["threshold"], 
                            left_subtree, right_subtree, best_split["varince_reduction"])
        
        # compute leaf node
        leaf_value = np.mean(Y)
        # return leaf node
        return TreeNode(value=leaf_value)
    
    def get_best_split(self, dset, num_samples, num_features):
        #function to find the best split

        # dictionary to store the best split
        best_split = {}
        max_var_red = -float("inf")# taking maximum value - infinity
        # looping across all the feature
        for feature_index in range(num_features):
            feature_values = dset[:, feature_index]
            possible_thresholds = np.unique(feature_values)
            # loop across all the values in the chosed feature
            for threshold in possible_thresholds:
                # split the dataset about threshold
                dataset_left, dataset_right = self.split(dset, feature_index, threshold)
                # checking if childs are not null 
                if len(dataset_left)>0 and len(dataset_right)>0:
                    y, left_y, right_y = dset[:, -1], dataset_left[:, -1], dataset_right[:, -1]
                    # compute information gain
                    curr_var_red = self.variance_reduction(y, left_y, right_y)
                    # update the best split if we found another maximum varience reduction 
                    if curr_var_red>max_var_red:
                        best_split["threshold"] = threshold
                        best_split["feature_index"] = feature_index
                        best_split["dataset_right"] = dataset_right
                        best_split["dataset_left"] = dataset_left
                        best_split["varince_reduction"] = curr_var_red
                        max_var_red = curr_var_red
        return best_split
    
    def split(self, dset, feature_index, threshold):
        #function to split the data 
        dataset_right = np.array([row for row in dset if row[feature_index]>threshold])
        dataset_left = np.array([row for row in dset if row[feature_index]<=threshold])
        return dataset_left, dataset_right
    
    def variance_reduction(self, parent, l_child, r_child):
        #compute variance reduction of a node 
        
        weight_l = len(l_child) / len(parent)
        weight_r = len(r_child) / len(parent)
        reduction = np.var(parent) - (weight_l * np.var(l_child) + weight_r * np.var(r_child)) # compute varience reduction 
        return reduction
    
    
                
    def print_tree(self,col, tree=None, indent=" "):
        #function to print the tree
        
        if not tree:
            tree = self.root

        if tree.value is not None:
            print(tree.value)

        else:
            print(str(col[tree.feature_index]), "<=", tree.threshold, "varince reduction", tree.varince_reduction)
            print("%sleft node :" % (indent), end="")
            self.print_tree(col,tree.left, indent + indent)
            print("%sright node :" % (indent), end="")
            self.print_tree(col,tree.right, indent + indent)
    
    
    
        
    def predict(self, X):
        #function to predict a single data point
        
        preditions = [self.modal_prediction(x, self.root) for x in X]
        return preditions
        
    def modal_prediction(self, x, tree):
        #function to predict new dataset
        
        if tree.value!=None: return tree.value
        feature_val = x[tree.feature_index]
        if feature_val<=tree.threshold:
            return self.modal_prediction(x, tree.left)
        else:
            return self.modal_prediction(x, tree.right)

```

```python
def train_test_split(x, y, test_size = 0.2, random_state = None):

	""" partioning the data into train and test sets """

	x_test = x.sample(frac = test_size, random_state = random_state)
	y_test = y[x_test.index]

	x_train = x.drop(x_test.index)
	y_train = y.drop(y_test.index)

	return x_train, x_test, y_train, y_test
```
